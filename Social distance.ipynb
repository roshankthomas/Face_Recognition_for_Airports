```python
import cv2
import warnings
warnings.filterwarnings('ignore')
```

```python



# Load the cascade
face_cascade = cv2.CascadeClassifier('HC_frontal_default.xml')
opc_count = 0
object_id_list = []
line_length_prev = 0
line_length_curr = 0
# To capture video from webcam. 
cap = cv2.VideoCapture(0)
# To use a video file as input 
# cap = cv2.VideoCapture('filename.mp4')
q = 0
while True:
    # Read the frame
    _, img = cap.read()
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect the faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    # Draw the rectangle around each face
    i = 0
    
    for (x, y, w, h) in faces:
        if w>110:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
            i = i+1
            if i not in object_id_list:
                object_id_list.append(i)
    opc_count = len(object_id_list)
    opc_text = ("Footfall:{}".format(opc_count))
    cv2.putText(img, opc_text,(5,30), cv2.FONT_HERSHEY_DUPLEX, 1, (255,255,255), 2)
            
    if len(faces)>1:        
        x1,y1,w1,h1 = faces[0]
        x2,y2,w2,h2 = faces[1]
        
        
        if w1 > 110 and w2> 110:
            
            
            line_length_prev = line_length_curr
            line_length_curr = int(((x1+(w1/2)-x2-(w2/2))**2 + (y1+(h1/2)-y2-(h2/2))**2)**0.5)
            if line_length_curr < line_length_prev/2:
                continue
            cv2.line(img,(x1+int(w1/2),y1+int(h1/2)),(x2+int(w2/2),y2+int(h2/2)),(255,0,0),3)
            min = (h2, h1) [h1 < h2] 
            max = (h1, h2) [h1 < h2] 
#             cv2.putText(img,"current distance between them : "+str(line_length_curr),(10,20),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
#             cv2.putText(img,'minor object height : '+str(min),(10,60),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
#             cv2.putText(img,'major object height : '+str(max),(10,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3)
            if line_length_curr < max:
                
                cv2.putText(img,'too close!!',(x1+6,y2-6),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),3)
                q=q+1
    cv2.putText(img,'Violation!'+str(q),(900,50),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),2)
                
                
            
    
            
                

    # Display
    cv2.imshow('img', img)
    # Stop if escape key is pressed
    k = cv2.waitKey(30) & 0xff
    if k==27:
        break
# Release the VideoCapture object
cap.release()
```
